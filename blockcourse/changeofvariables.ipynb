```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
```

```python
a = sps.norm(0,1).rvs(10000)
b = sps.norm(1,1).rvs(10000)
```

```python
plt.hist(a, bins = np.linspace(-5,5), histtype='step');
plt.hist(b, bins = np.linspace(-5,5), histtype='step');
```

```python
aa = sps.norm(0,1).cdf(a)
bb = sps.norm(0,1).cdf(b)
```

```python
plt.hist(aa, bins = np.linspace(0,1), histtype='step');
plt.hist(bb, bins = np.linspace(0,1), histtype='step');
```

```python
plt.scatter(a,aa, alpha = 0.2)
plt.scatter(b,bb, alpha = 0.2)
```

```python
import  seaborn as sns
sns.jointplot(a,aa)
plt.gcf().set_size_inches(5,5)
plt.gcf().savefig('cdf_transform')
```
