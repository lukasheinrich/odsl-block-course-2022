```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np
```

```python
mean = 3
variance = 4
n_sample = 10
samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

estimates = samples.mean(axis=1)
plt.hist(estimates, bins = 50, density = True);
plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');
plt.gcf().set_size_inches(5,5)
plt.vlines(mean,0,1.0,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.vlines(estimates.mean(),0,1.0,colors = 'cyan', linestyles='dashed', label = r'$E[\bar{x}]$')
plt.xlabel(r'sample mean $\bar{x}$')
plt.ylabel('density')
plt.title(fr'$\mu={mean},\;\sigma^2={variance},\;n={n_sample}$')
plt.legend()
plt.ylim(0,1)
plt.gcf().savefig('gaussian_mean.png')
```

```python
mean = 3
variance = 4
n_sample = 10
samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

for estimates,label in zip([
    samples[:,:1].mean(axis=1),
    samples.mean(axis=1)
],[r'$t(x) = x_0$', r'$t(x) = \bar{x}$']):
    plt.hist(estimates, bins = 50, density = True, label = label, alpha = 0.5);
    plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');
plt.gcf().set_size_inches(5,5)
plt.xlabel(r'estimator value')
plt.ylabel('density')
plt.title(fr'$\mu={mean},\;\sigma^2={variance},\;n={n_sample}$')
plt.legend()
plt.ylim(0,1)
plt.gcf().savefig('gauss_small_large_variance.png')
```

```python
for n_sample in [10,25,100]:
    mean = 3
    variance = 4
    samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

    estimates = samples.mean(axis=1)
    plt.hist(estimates, bins = 50, density = True, alpha = 0.5,label  = f'n={n_sample}');
    plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');

plt.gcf().set_size_inches(5,5)
plt.vlines(mean,0,2.5,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.xlabel(r'sample mean $\bar{x}$')
plt.ylabel('density')
plt.title(fr'$\mu={mean},\;\sigma^2={variance},\;n={n_sample}$')
plt.legend()
plt.ylim(0,2.5)
plt.gcf().savefig('gaussian_mean_consistency.png')
```

```python
mean = 3
variance = 4
n_sample = 10
samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

estimates = samples.var(axis=1, ddof=0)
plt.hist(estimates, bins = 50, density = True);
plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');
plt.gcf().set_size_inches(5,5)

plt.vlines(variance,0,0.5,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.vlines(estimates.mean(),0,0.5,colors = 'cyan', linestyles='dashed', label = r'$E[\bar{x}]$')


plt.vlines(4,0,0.4, color = 'k')
plt.vlines(4,0,0.4, color = 'k')
plt.ylim(0,0.4)
plt.xlabel('sample mean')
plt.ylabel('density')
plt.legend()
plt.gcf().savefig('gaussian_variance_uncorr.png')
```

```python
for n_sample in [10,25,100]:
    mean = 3
    variance = 4
    samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

    estimates = samples.var(axis=1, ddof=0)
    plt.hist(estimates, bins = 50, density = True, alpha = 0.5,label  = f'n={n_sample}');
    plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');

plt.gcf().set_size_inches(5,5)
plt.vlines(variance,0,2.5,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.xlabel(r'sample variance $s^2$')
plt.ylabel('density')
plt.title(fr'$\mu={mean},\;\sigma^2={variance},\;n={n_sample}$')
plt.legend()
plt.ylim(0,1.0)
plt.gcf().savefig('gaussian_variance_consistency.png')
```

```python
for n_sample in [10,25,100]:
    mean = 3
    variance = 4
    samples  = sps.norm(mean,np.sqrt(variance)).rvs(size  = (100000,n_sample))

    estimates = samples.var(axis=1, ddof=0)
    plt.hist(estimates, bins = 75, density = True, alpha = 0.5,label  = f'n={n_sample}');
    plt.hist(estimates, bins = 75, density = True, histtype='step', edgecolor  = 'k');
    b = plt.vlines(estimates.mean(),0,2.5,colors = 'r', linestyles='dashed', alpha = 0.5)
b.set_label(r'$E[s^2]$')
plt.xlim(1,7)
plt.gcf().set_size_inches(5,5)
plt.vlines(variance,0,2.5,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.xlabel(r'sample variance $s^2$')
plt.ylabel('density')
plt.title(fr'$\mu={mean},\;\sigma^2={variance},\;n={n_sample}$')
plt.legend()
plt.ylim(0,1.0)
plt.gcf().savefig('gaussian_variance_bias_disappears.png')
```

```python
estimates = samples.var(axis=1, ddof=1)
plt.hist(estimates, bins = 50, density = True);
plt.hist(estimates, bins = 50, density = True, histtype='step', edgecolor  = 'k');
plt.gcf().set_size_inches(5,5)

plt.vlines(variance,0,0.5,colors = 'k', linestyles='solid', label = r'$\theta_0$')
plt.vlines(estimates.mean(),0,0.5,colors = 'cyan', linestyles='dashed', label = r'$E[\bar{x}]$', zorder=999)

plt.vlines(4,0,0.4, color = 'k')
plt.vlines(4,0,0.4, color = 'k')
plt.ylim(0,0.4)
plt.xlabel('sample variance (corrected)')
plt.ylabel('density')
plt.legend()
```
