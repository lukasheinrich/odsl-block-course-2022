```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.arange(20)
p = sps.poisson(mu = 6).pmf(n)
```

```python
plt.bar(n,p, facecolor = 'maroon')
plt.bar(n[np.arange(3,10)],p[np.arange(3,10)], facecolor = 'grey')
plt.gcf().set_size_inches(5,5)
plt.ylim(0,.2)
plt.savefig('poisson_overcover.png')
```

```python
plt.bar(n,p, facecolor = 'maroon')
plt.bar(n[np.arange(2,11)],p[np.arange(2,11)], facecolor = 'grey')
plt.gcf().set_size_inches(5,5)
plt.ylim(0,.2)
plt.savefig('poisson_undercover.png')
```
