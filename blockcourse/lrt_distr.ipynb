```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = 2
toys = sps.norm(2).rvs(100000)
ti = np.linspace(0,40,10001)
stat = -2*(sps.norm.logpdf(toys,2) - sps.norm.logpdf(toys,toys))
plt.hist(stat,density=True, bins = 100, histtype = 'step', edgecolor = 'steelblue');
pi = sps.chi2(1).pdf(ti)
plt.plot(ti,pi, c = 'steelblue')


toys = sps.norm(0).rvs(100000)
stat = -2*(sps.norm.logpdf(toys,2) - sps.norm.logpdf(toys,toys))
plt.hist(stat,density=True, bins = 100, histtype = 'step', edgecolor = 'maroon');
pi = sps.ncx2(1,4).pdf(ti)


plt.plot(ti,pi, c = 'maroon')
plt.vlines(sps.ncx2(1,4).ppf([.5]),1e-5,1e3, colors = 'maroon')

plt.ylim(1e-5,1e3)
plt.semilogy()
```
