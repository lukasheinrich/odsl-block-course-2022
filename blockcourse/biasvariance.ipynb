```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
true = 0.0
bias = 5
sqrtvar = 1
est = np.random.normal(loc = true+bias, scale = sqrtvar, size = 100000)
mse = (est-true)**2
```

```python
sns.jointplot(est,mse, marginal_kws=dict(bins=100, rug=True),)
```

```python
plt.hist(mse,bins = 100, density=True);
plt.vlines(bias**2+sqrtvar**2,0,0.05, colors = 'r')
plt.vlines(mse.mean(),0,0.05, colors = 'k', linestyles='dashed')

```

```python
import numpy as np

#following  Kay Eldar https://msol.people.uic.edu/ECE531/papers/Rethinking%20Biased%20Estimation.pdf

true  = 6
var = 1*true**2
mi = np.linspace(-1,0)
newvar = (1+mi)**2*var
newbia2 = mi**2*true**2
newmse = newvar+newbia2

mstar = -1/(1+true**2/var)
plt.vlines(mstar,0,np.max(newmse))
plt.plot(mi,newvar, label =  'var')
plt.plot(mi,newbia2, label = 'bias')
plt.plot(mi,newmse, label =  'mse')
plt.legend()
```
