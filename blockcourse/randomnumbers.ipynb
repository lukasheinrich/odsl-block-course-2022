```python
import scipy.stats as sps
import matplotlib.pyplot as plt
x = sps.uniform(0,1).rvs(size = (50,))
print(x)
```

```python
dist = sps.multivariate_normal(mean = [2.0,3.0], cov = [[2.0,0.5],[0.5,1.2]]) 
x = dist.rvs(size = (250,))
print(x)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(5,5)})
sns.jointplot(x[:,0],x[:,1], marginal_kws=dict(bins =  31))
```

```python
x = np.random.normal(0, scale = 1,size = (10000,))
y  = x**2
sns.jointplot(x,y, marginal_kws=dict(bins = 100), joint_kws=dict(alpha = 0.2))
plt.gcf().set_size_inches(5,5)
plt.gcf().tight_layout()
plt.savefig('squash_chi2.png',dpi=600)
```

```python
xi = np.linspace(0,5,101)
yi = sps.chi2(df = 1).pdf(xi)
plt.hist(y, density=True, bins = np.linspace(0,5,101), label = 'samples')
plt.plot(xi,yi, label = r'$\chi^2_1$')
plt.xlabel(r'$y$')
plt.ylabel(r'$p(y)$')
plt.legend()
plt.gcf().set_size_inches(5,5)
plt.savefig('chi2_compare.png',dpi=600)
```

```python
for df in np.arange(5):
    xi = np.linspace(0,5,101)
    yi = sps.chi2(df = df).pdf(xi)
    plt.plot(xi,yi, label = f'k={df}')
plt.xlabel(r'$y$')
plt.ylabel(r'$p(y)$')
plt.legend()
plt.tight_layout()
plt.gcf().set_size_inches(5,5)
plt.savefig('chi2_various_df.png',dpi=600)
```
